```python
import enum
import attrs
from typing import Optional, Callable, Any, Literal
file_path='/Users/mattia/Desktop/SocialThings/Progetto/ImpressionMIND.csv'
import numpy as np
import scipy.sparse as sp
from scipy.sparse import csr_matrix
import pandas as pd
```

```python
dfmind=pd.read_csv(file_path, sep=',')
dfmind
```

```python
dfMIND = dfmind.sort_values(by='Timestamp')
dfMIND
```

```python
dfMIND['Impressions_Encoded'] = dfMIND['Impressions_Encoded'].apply(lambda x: eval(x))
dfMIND['History_encoded'] = dfMIND['History_encoded'].apply(lambda x: eval(x))
dfMIND
```

```python
usr_count = {}

for u in dfMIND['UserID'].unique():

    usr_count[u] = 0

usr_freq = []

for index, row in dfMIND.iterrows():

    u = row['UserID']

    usr_count[u] += 1

    usr_freq.append(usr_count[u])

dfMIND['UserFreq'] = usr_freq
```

```python
df_matrix = dfMIND.explode('Impressions_Encoded')[['ImpressionID','UserID', 'Impressions_Encoded', 'Time', 'Timestamp', 'UserFreq']]
df_matrix.reset_index(inplace=True,drop=True)
df_matrix
```

```python
from tqdm import tqdm

d = {}

for index, row in dfMIND.iterrows():

    id = row['ImpressionID']

    lst = row['Impressions_Encoded']

    d[id] = lst

positions = []

with tqdm(total=df_matrix.shape[0]) as pbar:

    for i in df_matrix.index:

        lst = d[df_matrix.iloc[i]['ImpressionID']]

        elem = df_matrix.iloc[i]['Impressions_Encoded']

        pos = lst.index(elem)

        positions.append(pos)

        pbar.update(1)

df_matrix['Position'] = positions
```

```python
import pandas as pd

# Creazione della colonna 'user-item'
df_matrix['user-item'] = df_matrix['UserID'].astype(str) + '-' + df_matrix['Impressions_Encoded'].astype(str)

# Creazione della colonna 'Conversion'
df_matrix['Conversion'] = df_matrix['Impressions_Encoded'].str.endswith('-1').astype(int)

# Creazione della colonna 'ImpCount'
df_matrix['ImpCount'] = df_matrix.groupby(['UserID', 'Impressions_Encoded']).cumcount() + 1

# Calcola la differenza di tempo tra la coppia corrente e l'ultima coppia user-item
df_matrix['LastSeen'] = df_matrix.groupby(['user-item'])['Timestamp'].diff()

# Imposta il valore 0 per la prima coppia user-item
df_matrix.loc[df_matrix.groupby(['user-item']).head(1).index, 'LastSeen'] = 0

# Trova le coppie con conversione uguale a 1
df_matrix['Conversion_Max'] = df_matrix.groupby(['user-item'])['Conversion'].cummax()

# Assegna un valore cumulativo per ogni riga del gruppo
df_matrix['GroupCumSum'] = df_matrix.groupby(['user-item'])['Conversion_Max'].cumsum()

# Filtra il dataframe in base alla condizione
new_df = df_matrix[df_matrix['GroupCumSum'] <= 1]

# Rimuovi le colonne aggiuntive
new_df = new_df.drop(columns=['Conversion_Max', 'GroupCumSum'])

# Converti i valori di LastSeen da secondi a giorni
new_df['LastSeen'] = (new_df['LastSeen'] / 86400)

# Seleziona le colonne desiderate nel nuovo dataframe
new_df = new_df[['user-item', 'ImpCount', 'Position', 'LastSeen', 'UserFreq', 'Conversion', 'Time']]

# Visualizza il nuovo dataframe
print(new_df)
```
